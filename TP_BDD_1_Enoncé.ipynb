## 1.2 Qu'est-ce qu'une base de données?

Une base de données peut être vue de différentes façons plus ou moins techniques :
- Une première approche est de voir une base de données comme un ensemble d'informations
modélisant les objets d'une partie du monde réel et servant de support à une application
informatique.
- Une base de données est au sens large une collection organisée, interrogeable et persistante
de données ou d'informations, concernant un thème donné pour l'entreprise.

D'un point de vue plus informatique, une base de données est :
- Un ensemble de fichiers comprenant un certain nombre d'articles formés chacun d'un type
de données, avec des liens entre ces articles.
- Un ensemble d'opérateurs permettant la recherche, le tri, la fusion ou toute autre opération
de même type.
- Un ensemble de données et de métadonnées. Ces dernières sont des données qui décrivent
les données, elles permettent en particulier de définir le schéma de la base.
Une base de données (ou database en anglais) est une entité dans laquelle il est possible de stocker
des données de façon structurée et avec le moins de redondance possible. Ces données doivent
pouvoir être utilisées par des programmes et des utilisateurs différents.

## 1.4 Les SGBD 

Un **Système de Gestion des Bases de Données** ou **SGBD** est l'outil principal de gestion
des bases de données.
C'est un ensemble de logiciels fournissant un environnement factorisant des modules de contrôle
des applications pour décrire, mémoriser, manipuler, traiter des ensembles de données, en
contrôler l'accès, les partager, gérer des pannes, . . . .

Comme le montre la figure ci-dessous, un SGBD permet aux utilisateurs via des applications de bases de données d'accéder aux informations de la base de données :
- Toutes les demandes ou requêtes des utilisateurs sont prises en charge efficacement par le
SGBD. C'est le SGBD qui, par exemple, s'occupe de fournir à l'utilisateur la liste des noms
de cinémas. L'utilisateur ne sait pas où se trouve l'information et comment l'extraire, c'est
le SGBD qui réalise ce travail.
- Il permet d'ajouter, d'effacer, de mettre à jour des données, et aussi de créer ou détruire
des fichiers, et d'accéder aux données. L'utilisateur pourra ainsi grâce au SGBD créer la
base de données cinéma et y saisir les informations concernant tous les cinémas.
Il permet de protéger l'utilisateur des détails du niveau matériel :
- L'utilisateur a une vue de la base à un niveau se situant au-dessus des niveaux matériels.
Le SGBD gère les demandes des utilisateurs. L'utilisateur a l'illusion que l'information est
comme il le souhaite.
- Chaque utilisateur a l'impression d'être le seul à utiliser la base de données.

>- l'en-tête : ensemble fixé d'**`attributs`**,
>- le corps : ensemble de **$t$-uplets**.

>Un **`t-uplet`** correspond à une ligne dans le corps de la relation, un **`attribut`** correspond à une colonne.

> Un **`domaine`** est un ensemble fini ou infini de valeurs admissibles pour un ou plusieurs attributs.

### 2.2.3 Quelques propriétés à respecter dans l'élaboration d'un schéma relationnel

Les propriétés décrites ci-dessous proviennent de la définition mathématique de la notion de **relation**. Dans cette théorie relationnelle, une relation est un **ensemble**.
<ol>
Les relations possèdent les <b>propriétés</b> suivantes :
    <li>pas de duplication de t-uplets (une relation est vue comme un ensemble qui ne peut contenir deux fois le même élément) </li>
    <li>les t-uplets ne sont pas ordonnés de haut en bas (les éléments est sans ordre)</li>
    <li>les attributs ne sont pas ordonnés de gauche à droite (changer l’ordre des colonnes ne change pas les t-uplets qui ne sont pas ordonnés) </li>
    <li>les valeurs d'un attribut font toutes parties d'un même domaine (chaque colonne associée à un attribut donné)</li>
    <li>les valeurs des attributs sont atomiques (Cf plus loin)</li>
 <hr />
Dans un schéma relationnel nous avons de plus les <b>propriétés</b> suivantes :
    <li>une relation a un nom unique dans le schéma</li>
    <li>un attribut dans une relation a un nom unique</li>
</ol>

Ces 7 propriétés sont explicites hormis la 5 ème qui doit être précisée: les valeurs des attributs sont **atomiques**. 
Que signifie ce terme **atomique** ?
Vous venez d'afficher le contenu de la table DISTRIBUTION qui possède comme attributs "Titre" et "NomActeur". Une autre proposition vous est faite ci-après.

Cette propriété provient du fait que les domaines contiennent des valeurs atomiques c'est à dire
**des valeurs qui ne peuvent pas être divisées en plusieurs sous-valeurs de même type**. Elle peut
aussi se traduire par le fait que lorsque l'on examine dans une relation l'intersection d'une ligne
et d'une colonne **il ne peut y avoir qu'une seule valeur, et non plusieurs**.

On a ici plusieurs noms d'acteurs pour un même film. Si on se permet de mettre ce genre d'information dans cette
relation il devient difficile de faire des recherches sur les acteurs d'un film, ou plutôt une partie des acteurs de ce film ou un seul acteur. Par exemple, on ne peut pas écrire dans ce cas directement une requête donnant les films ayant comme acteur "Freeman". 

Afin d'éviter cela, on répète autant de fois le même film que d'acteur nommé indépendamment comme dans votre base de données actuelle.

### 2.2.4 Clé primaire d'une relation 

Rappelons que le SGBD (système de gestion de base de données), comme `SQLite` que nous utilisons, permet de gérer une base de données dans sa globalité. Il doit ainsi assurer la cohérence de celle-ci au travers des contraintes suivantes:

- vérifier les données lors de leur chargement,
- vérifier les données lors de toute modification (saisie, mise à jour),
- répercuter certaines mises à jour entre tables,
- gérer les références entre tables.

Afin d'assurer cette mission, la définition d'une base de données est étendue par l'ajout de règles d'intégrité de natures différentes, dont l'une d'elle, est la suivante:
- *l'unicité de clé : Dans une base de données, toutes les relations doivent posséder une clé unique, appelée clé primaire*

En réalité, nous avons déjà évoqué cette règle sans parler de clé dans les propriétés des relations au paragraphe précédent: **pas de duplication des t-uplets (propriété 1)**. Pour qu'il n'y ait pas d'ambiguïté et que la cohérence de la base de données soit complète auprès du SGBD, **il est indispensable que les t-uplets soient uniques et qu'il n'y ait pas de redondance au niveau des lignes d'une même table**.


Si on reprend la relation CINEMA, il faut trouver un moyen qui garantisse la distinction des t-uplets (ligne de la table) entre eux. Une solution est de choisir un, ou une combinaison des attributs qui puissent le permettre. Par exemple, l'attribut NumeroCinema permet cette distinction univoque des t-uplets entre eux.

```python
%sql SELECT * FROM cinema
```

En effet, chaque ligne est numérotée. Ce numéro est unique. 

Rappelons le modèle relationnel de la base de données CINEMAS.

On comprend désormais le symbole de clé avec l'abrégé **CP** : l'attribut NumeroCinema de la table Cinema est la **Clé Primaire** de la table. Rien qu'en se limitant à cet attribut, on est capable de distinguer sans ambiguïté les t-uplets de cette relation (table). <br>
<u>Attention</u>: cet abrégé CP n'est pas normalisé. Vous pouvez trouver la clé primaire en gras précédée d'un astérisque.

En utilisant le SGBD SQLite, on se propose de modifier cette table en rajoutant un nouveau cinema qui prendra le numéro 1,  le nom de Vox et qui se situera à l'île de La Réunion. Voyons ce que répond le SGBD après l'instruction suivante qui est à valider!

## 2.2.5 D'autres règles d'intégrité 

Nous venons de voir en détail la notion de clé primaire qui permet d'assurer l'unicité des t-uplet (ligne) dans une même relation. Comme déjà évoqué dans un paragraphe précédent, cette règle fait partie d'un ensemble de règles nommées **règles d'intégrité**.

D'autres règles d'intégrité, en plus de celle-ci, existent et doivent être surveillées par le SGBD:
1. Unicité de clé
2. les contraintes de domaine, qui permettent de vérifier l'intégrité des attributs,
3. la règle d'intégrité des entités qui permet de préciser quelles valeurs doivent obligatoirement être définies ($HP$)
4. et enfin les contraintes de références, qui font appel aux notions de clés étrangères et qui permettent de vérifier la cohérence des informations entre différentes relations.

La seconde règle concerne les contraintes de domaine: Chaque attribut d'une relation prend ses valeurs dans un domaine, dont nous avons vu qu'il
s'agissait d'un ensemble. Dans la pratique, les domaines correspondent souvent à des types de base tels que : entier, réel, chaine de caractères, . . .
Si l'on considère, par exemple un attribut tel que CodePostalCinema, dans la relation CINEMA, on constate que le domaine de cet attribut est l'ensemble des entiers (compris entre 1000 et 99999).
Une contrainte de domaine est une contrainte d'intégrité qui définit la propriété que doit vérifier toute valeur d'un attribut d'une relation donnée.
Ce type de contrainte, traité par le SGBD, permet lors de la saisie ou la mise à jour d'information, de vérifier la cohérence des données.

## 3.1 La requête de sélection de données 

Nous avons plusieurs fois utilisé cette requête: 

    SELECT * FROM cinema

Cette requête très simple permet d'afficher (ou sélectionner) toute la relation (table) CINEMA de la base de données cinemas: attributs et t-uplet.

De manière générale, l'obtention de données se fait exclusivement avec l'ordre SELECT. Cet ordre possède 6 clauses
différentes dont seules les deux premières sont obligatoires: SELECT et FROM, comme précédemment.

L'évaluation d'une requête complète est réalisée dans l'ordre suivant

**Syntaxe de l'ordre SELECT**
1. `SELECT Liste-Attributs` (liste-attributs=nom des attributs à sélectionner) ou `SELECT DISTINCT` ou `SELECT Liste-Attributs AS` (voir plus bas)

**Ordre d'exécution des clauses:** 

2. `FROM Liste-tables` (Liste-tables=noms des tables utilisées)
3. `WHERE Condition` (Condition=contrainte imposée)

On verra plus tard d'autres clauses :

5. `GROUP BY Expression` (fonctions d'agrégation)
6. `HAVING Condition` (condition sur un groupe)
7. `ORDER BY Expression ASC/DESC` (trier les résultats par ordre croissant / décroissant)
8. `LIMIT n` (limiter à n enregistrements)
9. `OFFSET n` (débuter à partir de n enregistrements)
10. `UNION | INTERSECT | EXCEPT` opérations ensemblistes sur les requêtes

**Produit cartésien**: le produit cartésien permet de définir une nouvelle relation consistant en l'ensemble de tous
les t-uplets résultant de la combinaison de 2 t-uplets provenant des 2 relations considérées.

On peut bien entendu ne sélectionner en sortie que les attributs NomCinema et Titre avec la requête suivante. Mais le produit cartésien des 2 relations cinema et titre s'effectue bien à partir des clés de chacune.

```python
%sql SELECT NomCinema, Titre FROM cinema, film
```

Pour éviter les confusions losqu'on combine des attributs définis dans des relations différentes, on a vu précédemment que nous pouvions préciser les attributs en lui associant la relation dans laquelle elle est définie. La requête suivante est identique à la précédente.

```python
%sql SELECT cinema.NomCinema, film.Titre FROM cinema, film
```
